```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
X = boston.data
y = boston.target
```

```python
#normalise
u = np.mean(X,axis=0)
sd = np.mean(X,axis=0)
X = (X-u)/sd
```

```python
pd.DataFrame(X[:10,:])
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
X.drop(columns=[])
```

```python
def hypothesis(X,theta):
    y_ = np.dot(X,theta)
    return y_

def gradient(X,y,theta):
    m,n = X.shape
    y_ = hypothesis(X,theta)
    grad = np.dot(X.T,(y_ - y))
    return grad/m

def error(X,y,theta):
    m = X.shape[1]
    y_ = hypothesis(X,theta)
    total_error = np.sum((y_ - y)**2)
    return total_error/m

def gradient_descent(X,y,max_iters = 5000,learning_rate = 0.1):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    for i in range(max_iters):
        grad = gradient(X,y,theta)
        y_ = hypothesis(X,theta)
        e = error(X,y,theta)
        error_list.append(e)
        theta = theta - learning_rate*grad
        
    return theta,error_list
```

```python
theta,error_list = gradient_descent(X,y)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.plot(error_list)
plt.show()
```

```python
#score of model
def r2_score(y,y_):
    num = np.sum((y-y_)**2)
    denom = np.sum((y-y.mean())**2)
    score = (1 - num/denom)
    return score*100
```

```python
y_ = hypothesis(X,theta)
y_
```

```python
r2_score(y,y_)
```
